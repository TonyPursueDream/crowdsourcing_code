Check user's frequent visit venues

```python
def user_frequent_visit_locations(df, user_id):
    user_df = df[df.user_id == user_id]
    ret_df = user_df.groupby(['venue_id', 'venue_category_name']).size().reset_index(name='counts')
    ret_df = ret_df.sort_values(by=['counts'])
    return ret_df
```

```python
df_470 = user_frequent_visit_locations(nyc_df, 470)
df_470
```

```python
def general_frequent_visited_locations(df):
    ret_df = df.groupby(['venue_id', 'venue_category_name']).size().reset_index(name='counts')
    ret_df = ret_df.sort_values(by=['counts'])
    return ret_df
```

```python
general_frequent_visited_locations(nyc_df)
```

```python
def general_frequent_visited_categories(df):
    ret_df = df.groupby(['venue_category_name']).size().reset_index(name='counts')
    ret_df = ret_df.sort_values(by=['counts'])
    return ret_df
```

```python
g = general_frequent_visited_categories(nyc_df)
pd.set_option('display.max_rows', g.shape[0]+1)
print(g)
```
